```python
# ================================
# Corporate Strategy Analysis: Coca-Cola
# ================================

import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud

sns.set_style("whitegrid")

# -------------------------------
# Section 1: Introduction
# -------------------------------

print("# Corporate Strategy Analysis: Coca-Cola")
print("## Overview")
print("The Coca-Cola Company is a global beverage leader operating in more than 200 countries.")
print("This analysis explores its corporate strategy using financial, competitive, and strategic perspectives.")

# -------------------------------
# Section 2: Financial Analysis
# -------------------------------

ko = yf.Ticker("KO")

# Historical stock data (5 years)
hist = ko.history(period="5y")

plt.figure(figsize=(10,5))
sns.lineplot(data=hist, x=hist.index, y="Close", color="red")
plt.title("Coca-Cola Stock Price Trend (5 Years)")
plt.xlabel("Year")
plt.ylabel("Stock Price (USD)")
plt.show()

# -------------------------------
# Income Statement Analysis
# -------------------------------

income_stmt = ko.get_income_stmt()
income_stmt = income_stmt.T


# Extract Revenue
revenue = income_stmt["TotalRevenue"].dropna().sort_index()

plt.figure(figsize=(10,5))
sns.lineplot(x=revenue.index, y=revenue.values/1e9, marker="o", color="darkred")
plt.title("Coca-Cola Total Revenue (USD Billions)")
plt.xlabel("Year")
plt.ylabel("Revenue (Billion USD)")
plt.show()

# CAGR Calculation
start_value = float(revenue.iloc[0])
end_value = float(revenue.iloc[-1])
years = len(revenue) - 1

cagr = (end_value / start_value) ** (1 / years) - 1

print(f"\nRevenue CAGR (historical): {cagr:.2%}")

# -------------------------------
# Section 3: Porter’s Five Forces
# -------------------------------

forces = {
    "Supplier Power": 3,
    "Buyer Power": 4,
    "Industry Rivalry": 8,
    "Threat of Substitutes": 7,
    "Threat of New Entrants": 2
}

plt.figure(figsize=(8,5))
sns.barplot(x=list(forces.keys()), y=list(forces.values()), palette="Reds")
plt.title("Porter’s Five Forces – Coca-Cola")
plt.ylabel("Impact Level (1-10)")
plt.xticks(rotation=20)
plt.show()

# -------------------------------
# Section 4: SWOT Analysis Word Cloud
# -------------------------------

swot = {
    "Strengths": "Global brand equity Strong distribution network Pricing power High margins",
    "Weaknesses": "Dependence on carbonated drinks Sugar exposure Regulatory pressure",
    "Opportunities": "Emerging markets Premiumization Health beverages ESG positioning",
    "Threats": "Sugar taxes Commodity inflation FX volatility Competitive pressure"
}

wordcloud = WordCloud(width=800, height=400, background_color='white').generate(" ".join(swot.values()))

plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("SWOT Analysis – Coca-Cola")
plt.show()

# -------------------------------
# Section 5: Strategic Recommendations
# -------------------------------

recommendations = [
    "1. Accelerate low-sugar and functional beverage expansion.",
    "2. Strengthen presence in high-growth emerging markets.",
    "3. Increase premium product portfolio to enhance margins.",
    "4. Optimize supply chain through digitalization and AI."
]

print("\n## Strategic Recommendations")
for rec in recommendations:
    print(f"- {rec}")
```
